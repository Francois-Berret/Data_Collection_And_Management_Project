```python
#Spider pour visualiser la particularité des hébergements dont le scraping des détails n'a pas été possible depuis leur url

import os
import logging
import scrapy
from scrapy.crawler import CrawlerProcess

class BookingCheckSpider(scrapy.Spider):
    name = "BookingCheck"

    start_urls = off['url']
    download_delay = 1

    def start_requests(self):
        for url in self.start_urls:
            yield scrapy.Request(url=url, callback=self.scrape_results)


    def scrape_results(self, response):
        if response.xpath('//*[@id="js--hp-gallery-scorecard"]/a/div/div/div/div[1]/text()').get() is None:
            has_rating = False
        else:
            has_rating = response.xpath('//*[@id="js--hp-gallery-scorecard"]/a/div/div/div/div[1]/text()').get()

        yield {
            'url': response.url,
            'has_rating': has_rating
        }

filename = "Off_Hebergements.json"

if filename in os.listdir('/content/drive/MyDrive/Jedha_Fullstack/Kayak_Project/'):
        os.remove('/content/drive/MyDrive/Jedha_Fullstack/Kayak_Project/' + filename)

process = CrawlerProcess(settings = {
    'USER_AGENT': 'Chrome/97.0',
    'LOG_LEVEL': logging.INFO,
    'DOWNLOAD_DELAY' : 1,
    "FEEDS": {
        '/content/drive/MyDrive/Jedha_Fullstack/Kayak_Project/' + filename: {"format": "json"},
    }
})

process.crawl(BookingCheckSpider)
process.start()
```

```python
#Réalisation de la carte des meilleurs hébergements selon Booking pour chacune des 35 destinations

df_details['lat'] = df_details['geos'].apply(lambda x: x.split(',')[0]).astype(float)
df_details['lon'] = df_details['geos'].apply(lambda x: x.split(',')[1]).astype(float)
df_details['note'] = df_details['note'].apply(lambda x: x.replace(",",".")).astype(float)
df_details.head(3)
```

```python
import plotly.express as px

fig = px.scatter_mapbox(df_details, lat="lat", lon="lon", hover_name='hebergement', color="note",
                        mapbox_style="open-street-map", range_color=[5, 10], color_continuous_scale=px.colors.sequential.Inferno, zoom=4.5)

fig.update_traces(marker=dict(size=15))
fig.update_layout(title='Meilleurs hébergements par destination selon Booking')

fig.show()
```

```python
#Réalisation de la carte des meilleurs hébergements selon Booking pour les 5 destinations présentant le score subjectif météo médian le plus élevé pour les 5 jours à venir

top5Weathers = pd.read_csv('/content/drive/MyDrive/Jedha_Fullstack/Kayak_Project/Top5Weathers')
top5Weathers
```

```python
mask = (df_details['city'] == top5Weathers['ville'][0]) | (df_details['city'] == top5Weathers['ville'][1]) | (df_details['city'] == top5Weathers['ville'][2]) | (df_details['city'] == top5Weathers['ville'][3]) | (df_details['city'] == top5Weathers['ville'][4])
df5 = df_details.loc[mask,:]
```

```python
fig = px.scatter_mapbox(df5, lat="lat", lon="lon", hover_name='hebergement', color="note",
                        mapbox_style="open-street-map", range_color=[5, 10], color_continuous_scale=px.colors.sequential.Inferno, zoom=7)

fig.update_traces(marker=dict(size=15))
fig.update_layout(title='Meilleurs hébergements selon Booking des 5 destinations où il fera le plus beau dans les 5 jours à venir')

fig.show()
```
