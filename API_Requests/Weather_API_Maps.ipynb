```python
import pandas as pd
import requests
import json
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
#Liste des 35 villes à étudier

cities = ["Mont Saint Michel",
"St Malo",
"Bayeux",
"Le Havre",
"Rouen",
"Paris",
"Amiens",
"Lille",
"Strasbourg",
"Chateau du Haut Koenigsbourg",
"Colmar",
"Eguisheim",
"Besancon",
"Dijon",
"Annecy",
"Grenoble",
"Lyon",
"Gorges du Verdon",
"Bormes les Mimosas",
"Cassis",
"Marseille",
"Aix en Provence",
"Avignon",
"Uzes",
"Nimes",
"Aigues Mortes",
"Saintes Maries de la mer",
"Collioure",
"Carcassonne",
"Ariege",
"Toulouse",
"Montauban",
"Biarritz",
"Bayonne",
"La Rochelle"]
```

```python
#Test de la réponse à la requête API "coordonnées géographiques" sur une ville dans un premier temps

r = requests.get('https://nominatim.openstreetmap.org/search?q=bayonne+france&format=json')
r
```

```python
#Observation du résultat sous format JSON

r.json()[:2]
```

```python
#Réalisation de la requête sur l'ensemble des villes de l'étude, avec réessai en cas d'échec

import time

villes = []

for ville in cities:
    max_retries = 3
    retries = 0

    while retries < max_retries:
        try:
            r = requests.get("https://nominatim.openstreetmap.org/search?q={}+france&format=json".format(ville))
            r.raise_for_status()
            villes.append(r.json())
            break  #Sortie de boucle si la requête réussit
        except (requests.RequestException, ValueError) as e:
            retries += 1
            print("Une erreur s'est produite pour {}: {}. Réessai en cours...".format(ville, str(e)))
            time.sleep(1)  #1 seconde avant de réessayer

villes[:2]
```

```python
#Récupération des coordonnées géographiques des 35 villes

villes = [villes[i][0] for i in range(len(villes))]
```

```python
#Observation du résultat format de la requête au format JSON pour les 35 villes

villes[:4]
```

```python
#Conversion en dataframe

df_villes = pd.DataFrame(villes)
df_villes.head()
```

```python
df_villes.columns
```

```python
#Dataframe des coordonnées géographiques des 35 villes

df_villes_xy = df_villes.loc[:,["display_name", "lat", "lon"]]
df_villes_xy
```

```python
#Test de la réponse à la requête API "météo" sur une ville dans un premier temps

r2 = requests.get("https://api.openweathermap.org/data/2.5/forecast?lat=43.21&lon=2.34&units=metric&appid=___")
r2.json()['city']
```

```python
#Récupération des données météorologiques des 35 villes

cities_lat = df_villes_xy["lat"]
cities_lon = df_villes_xy["lon"]
cities_weather = []
for lat, lon in zip(cities_lat, cities_lon):
    r = requests.get("https://api.openweathermap.org/data/2.5/forecast?lat={}&lon={}&units=metric&appid=___".format(lat,lon))
    cities_weather.append(r.json())
```

```python
#OpenWeatherMap ne renvoie pas les noms des monuments, sites naturels ou départements de la liste initiale mais les villes associées à ces lieux

owm_cities_names = []
for i in range(len(cities_weather)):
   owm_cities_names.append(cities_weather[i]['city']['name'])
owm_cities_names
```

```python
#Modification des noms des villes pour concordance entre données API GPS et données API météo (OpenWeatherMap arrondit les données GPS ce qui peut changer l'intitulé exact d'une ville)

df_villes_xy["display_name"] = owm_cities_names
df_villes_xy = df_villes_xy.rename(columns={"display_name":"ville"})
df_villes_xy
```

```python
#Création d'un dataframe avec les données météo sélectionnées pour les 35 villes

WeatherData = pd.DataFrame()

for city_weather in cities_weather:
    city = city_weather['city']['name']
    city_data = []  #Liste temporaire pour stocker les données de la ville actuelle

    for obs in range(city_weather['cnt']):
        nouvelle_ligne = {'ville': city,
                          'date': city_weather["list"][obs]['dt_txt'],
                          'temp_min': city_weather["list"][obs]['main']['temp_min'],
                          'temp_max': city_weather["list"][obs]['main']['temp_max'],
                          'description': city_weather['list'][obs]['weather'][0]['description'],
                          'clouds_score': city_weather['list'][obs]['clouds']['all'],
                          'pop': city_weather['list'][obs]['pop']
                          }

        city_data.append(nouvelle_ligne)

    city_df = pd.DataFrame(city_data)  #DataFrame pour les données de la ville actuelle
    WeatherData = pd.concat([WeatherData, city_df], ignore_index=True)

WeatherData
```

```python
#Dataframe final des données météo sélectionnées pour chacune des 35 villes avec rétablissement des noms des 35 lieux de la liste initiale

WeatherData = pd.merge(WeatherData, df_villes_xy, on="ville", how="inner")

mapping_dict = dict(zip(WeatherData['ville'].unique(), df_villes['display_name'].unique()))
WeatherData['ville'] = WeatherData['ville'].replace(mapping_dict)
WeatherData['ville'] = WeatherData['ville'].str.split(',').str[0].str.strip()

WeatherData
```

```python
#Encodage ordinal des catégories de "ciels"

WeatherData['description'].unique()
```

```python
WeatherData['description_score'] = WeatherData['description'].replace({"heavy intensity rain" : 1, "moderate rain" : 2, "light rain" : 3,
                                                                       "overcast clouds" : 4, "broken clouds" : 5, "scattered clouds" : 6,
                                                                       "few clouds" : 7, "clear sky" : 8})
```

```python
#Détermination d'un "score" météo subjectif

WeatherData['score_meteo_subjectif'] = WeatherData['description_score']*2 + WeatherData['temp_max']
WeatherData
```

```python
WeatherData.to_csv('/content/drive/MyDrive/Jedha_Fullstack/Kayak_Project/WeatherDataframe', index=False)
```

```python
WeatherData.dtypes
```

```python
WeatherData["lat"] = WeatherData["lat"].astype(float)
WeatherData["lon"] = WeatherData["lon"].astype(float)
```

```python
#Carte de l'évolution du score météo subjectif des 35 villes dans les 5 jours à venir

import plotly.express as px

fig = px.scatter_mapbox(WeatherData, lat="lat", lon="lon", color='score_meteo_subjectif', hover_name='ville',
                        mapbox_style="open-street-map", color_continuous_scale=px.colors.sequential.thermal, zoom=3.8,
                       size_max=30, animation_frame='date')

fig.update_traces(marker=dict(size=16))
fig.update_layout(coloraxis_colorbar=dict(title='Score meteo subjectif'))

fig.show()
```

```python
#Top-5 villes en termes de score météo médian pour les 5 jours à venir

top_5 = WeatherData.groupby("ville")["score_meteo_subjectif"].median().sort_values(ascending=False)[:5]
top_5
```

```python
#DataFrame des top-5 villes

import re

lat = []
lon = []

for ville in top_5.index:
    lat.append(WeatherData.loc[WeatherData['ville'] == ville, ['lat']].max())
    lon.append(WeatherData.loc[WeatherData['ville'] == ville, ['lon']].max())

lat_values = [float(re.findall(r"\d+\.\d+", str(value))[0]) for value in lat]
lon_values = [float(re.findall(r"\d+\.\d+", str(value))[0]) for value in lon]
```

```python
top_5_df = pd.DataFrame({"ville" : top_5.index, "lat" : lat_values, "lon" : lon_values, "top_scores_medians_j+5" : top_5.values})
top_5_df
```

```python
top_5_df.to_csv('/content/drive/MyDrive/Jedha_Fullstack/Kayak_Project/Top5Weathers', index=False)
```

```python
#Carte des top-5 villes en termes de météo pour les 5 jours à venir

fig = px.scatter_mapbox(top_5_df, lat="lat", lon="lon", hover_name='ville', color="top_scores_medians_j+5",
                        mapbox_style="open-street-map", color_continuous_scale=px.colors.sequential.Oryel, zoom=5)

fig.update_traces(marker=dict(size=30))
fig.update_layout(coloraxis_colorbar=dict(title='Score meteo subjectif médian J+5'))

fig.show()
```
